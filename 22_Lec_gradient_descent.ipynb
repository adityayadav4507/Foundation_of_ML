```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#generate simple dataset of y=2x+1 with noise
np.random.seed(42)
X=np.linspace(0,10,50) # 50 evenly distributed points between 0 to 10
true_m, true_c = 2,1
Y=true_m*X+true_c+np.random.normal(0,2,X.shape) #Add some noise

```

```python
#initalize the parameter slope and intercept
m,c=0.0,0.0
ler_rate=0.01 #learnining rate
iterations=100 #number of iterations

```

```python
#store losses
losses_history = []
```

# For Gradient Descent

```python
for i in range(iterations):
  # y predicted
  y_pre=m*X+c
  error=y_pre-Y

  #computed gradent

  m_grad=2/len(X) *np.sum(X*error)
  c_grad=2/len(X) *np.sum(error)

 # update parameter

  m -= ler_rate*m_grad
  c -= ler_rate*c_grad

  # compute losses

  loss=np.mean(error**2)
  losses_history.append(loss)


  #print updated parameter
  if i % 10 ==0:
   print("itereation=", i , " losses=", loss , "m=", m , "c=", c)

```

```python
# final parameter
print("m=",m , "c=",c )
```

```python
# plot the dataset of fitted line
plt.figure(figsize=(10,6))
plt.scatter(X,Y,label="datapoint" , color="blue")
plt.plot(X,m*X+c , label="Fitted Line"  , color="red")
plt.title("Gradent Descent Fitted Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# plot loss history
plt.figure(figsize=(10,6))
plt.plot(range(iterations), losses_history, label="loss")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```
